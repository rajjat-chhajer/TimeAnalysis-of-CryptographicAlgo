```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
from Crypto.Random import get_random_bytes
from datetime import datetime
import pandas as pd
timeanalysis= pd.DataFrame([])
for _ in range(0,100):
    start = datetime.now()
    # Generate RSA key pair with a 1024-bit key
    key = RSA.generate(1024)

    # Get the public key for encryption
    public_key = key.publickey()

    # Get the private key for decryption
    private_key = key

    # Message to be encrypted
    message = get_random_bytes(16)

    # Encrypt the message using the public key
    cipher = PKCS1_OAEP.new(public_key)
    encrypted_message = cipher.encrypt(message)

    # Decrypt the message using the private key
    cipher = PKCS1_OAEP.new(private_key)
    decrypted_message = cipher.decrypt(encrypted_message)
    end = datetime.now()
    elapsetime = (end - start).total_seconds()
    timeanalysis = timeanalysis.append([elapsetime])
```

```python
print('maximum amount of time for one exchange',timeanalysis.values.max())
print('minimum amount of time for one exchange',timeanalysis.values.min()) 
print('Average amount of time for one exchange', timeanalysis.values.mean())
```
